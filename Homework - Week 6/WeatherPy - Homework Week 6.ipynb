```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lat = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lat, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
# OpenWeatherMap API Key
api_key = api_keys.api_key

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 
```

```python
#Create list where city data will be collected
city_data = []

#Loop through cities to iterate the city from each row for API call using the enumerate function
for i, city in enumerate(cities):
        
    #Variable created to piece the URL together for each city in our random list as we loop through
    CityURL = url + "&q=" + city
    
    #API request
    try:
        #Use Request function to pull in the JSON request
        CityWeather = requests.get(CityURL).json()

        #Pull out the data we need from the directories in the JSON request
        #Example: http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=ca7e42fee6a31584451ea76e79b3bf19&q=chicago
        #{"coord":{"lon":-87.62,"lat":41.88},
        #"weather":[{"id":701,"main":"Mist","description":"mist","icon":"50n"}],
        #"base":"stations",
        #"main":{"temp":74.07,"pressure":1016,"humidity":94,"temp_min":71.6,"TempMax":75.2},
        #"visibility":8047,
        #"wind":{"speed":5.73,"deg":277.501},
        #"clouds":{"all":40},"dt":1533611700,
        #"sys":{"type":1,"id":1030,"message":0.0045,"country":"US","sunrise":1533639031,"sunset":1533690073},
        #"id":4887398,
        #"name":"Chicago",
        #"cod":200}
        clng = CityWeather["coord"]["lon"]
        clat = CityWeather["coord"]["lat"]
        chumidity = CityWeather["main"]["humidity"]
        cmax_temp = CityWeather["main"]["TempMax"]
        cwind = CityWeather["wind"]["speed"]
        cclouds = CityWeather["clouds"]["all"]
        ccountry = CityWeather["sys"]["country"]
        cdate = CityWeather["dt"] 

        #Use append to add the city data to a list
        city_data.append({"City": city, 
                          "Lat": clat, 
                          "Lng": clng, 
                          "Max Temp": cmax_temp,
                          "Humidity": chumidity,
                          "Clouds": cclouds,
                          "Wind": cwind,
                          "Country": ccountry,
                          "Date": cdate})
    except:
        print("No city match")

#JSONS to Pandas dataframe conversion
city_data_pd = pd.DataFrame(city_data)

#Fields needed for graphs
lats = city_data_pd["Lat"]
max_temps = city_data_pd["Max Temp"]
humidity = city_data_pd["Humidity"]
Clouds = city_data_pd["Clouds"]
wind_speed = city_data_pd["Wind"]

# Export the City_Data into a csv
city_data_pd.to_csv(output_data_file, index_label="City_ID")

#City DataFrame display to confirm loop worked
city_data_pd.head()
```
